```python
#Visualise bart
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

pd_file = pd.read_csv('bart.csv')
pd_file
```

```python
pd_file = pd_file.as_matrix()
```

```python
#Gather data for 2st plot
mean_after_no_explosion = []
mean_after_explosion = []
for row in pd_file:
    mean_after_explosion.append(row[9])
    mean_after_no_explosion.append(row[10])
```

```python
#Plot mean after explosion vs mean after no explosion
plt.plot(mean_after_explosion, 'c', label='Mean after explosion')
plt.plot(mean_after_no_explosion, 'm', label='Mean after no explosion')
plt.margins(0.05, 0.1)
plt.legend(bbox_to_anchor=(1.05,1), loc=2)
plt.xlabel('Attempt (#)')
plt.ylabel('Mean # of Pumps')
plt.show()
```

```python
#Gather data for 2st plot
mean_pumps = []
num_of_explosion = []
for row in pd_file:
    num_of_explosion.append(row[5])
    mean_pumps.append(row[6])
```

```python
#Plot 2nd plot
plt.plot(num_of_explosion, 'c', label='Number of Explosions')
plt.plot(mean_pumps, 'm', label='Mean no. of Pumps')
plt.margins(0.05, 0.1)
plt.legend(bbox_to_anchor=(1.05,1), loc=2)
plt.xlabel('Attempt (#)')
plt.show()
```
